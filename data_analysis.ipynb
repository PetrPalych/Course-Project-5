```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Загрузка данных

```python
df = pd.read_csv("basel(1940-2023).csv")
```

```python
sns.histplot(data=df['Temperature [2 m elevation corrected]'], bins=30, kde=True, color='skyblue', edgecolor='blue', linewidth=1)
```

```python
columns_list = ['Temperature [900 mb]', 'Temperature [850 mb]', 'Temperature [800 mb]', 'Temperature [700 mb]', 'Temperature [500 mb]']

rows = 2
cols = (len(columns_list) + rows - 1) // rows

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(12, 4 * rows))

for i, column in enumerate(columns_list):
    ax = axes[i // cols, i % cols]
    sns.boxplot(data=df, y=column, x='Month', palette='Set2', ax=ax)
    ax.set_title(f'Boxplot for {column}')

fig.delaxes(axes[1, 2])

fig.suptitle('Boxplots for Different Temperatures', y=1.02)
plt.tight_layout()
```

На графике видно, чем ниже атмосферное давление, тем ниже температура. Вне зависимости какой бы не был бы месяц. Хотя первые три графика в периоде с 6 по 9 большой разницы не имеют.

```python
columns_list = ['Wind Speed [900 mb]', 'Wind Speed [850 mb]', 'Wind Speed [800 mb]', 'Wind Speed [700 mb]', 'Wind Speed [500 mb]', 'Wind Speed [250 mb]']

rows = 2
cols = (len(columns_list) + rows - 1) // rows

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(12, 4 * rows))

for i, column in enumerate(columns_list):
    ax = axes[i // cols, i % cols]
    sns.boxplot(data=df, y=column, x='Month', palette='Set2', ax=ax)
    ax.set_title(f'Boxplot for {column}')

fig.suptitle('Boxplots for Different Wind Speeds', y=1.02)
plt.tight_layout()
```

```python
geopotential_data = df[['Year', 'Month', 'Hour', 'Geopotential Height [1000 mb]', 'Geopotential Height [850 mb]',
                         'Geopotential Height [800 mb]', 'Geopotential Height [700 mb]', 'Geopotential Height [500 mb]']]

plt.figure(figsize=(12, 8))

sns.lineplot(x='Hour', y='Geopotential Height [1000 mb]', data=geopotential_data, label='1000 mb')
sns.lineplot(x='Hour', y='Geopotential Height [850 mb]', data=geopotential_data, label='850 mb')
sns.lineplot(x='Hour', y='Geopotential Height [800 mb]', data=geopotential_data, label='800 mb')
sns.lineplot(x='Hour', y='Geopotential Height [700 mb]', data=geopotential_data, label='700 mb')
sns.lineplot(x='Hour', y='Geopotential Height [500 mb]', data=geopotential_data, label='500 mb')

plt.title('Geopotential height at different levels')
plt.ylabel('Geopotential height (Gpm)')
plt.legend(title='Atmospheric pressure level')
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x='Wind Direction [10 m]', y='Wind Speed [10 m]', data=df, hue='Month', alpha=0.2)

plt.title('Scatter diagram for wind')
plt.xlabel('Wind direction (°)')
plt.ylabel('Wind speed (m/s)')
```

Диаграмму разброса при взаимосвязи направления и скорости ветра никак не изменяется(при Wind Direction [100 m] и Wind Speed [100 m), отсюда следует, что погодные условия стабильны. Первые месяцы скорость ветра высока, в середине ниже всего, а под конец года снова поднимается. Направление ветра никак не влияет на месяц года.

```python
plt.figure(figsize=(12, 6))

levels = ['900 mb', '850 mb', '800 mb', '700 mb', '500 mb', '250 mb']

for level in levels:
    sns.lineplot(x='Month', y=f'Wind Direction [{level}]', data=df, label=f'{level}')

plt.title('Graph of wind direction at different levels')
plt.ylabel('Wind direction (°)')
plt.legend(title='Level')
```

```python
plt.figure(figsize=(10, 6), dpi=200)

sns.barplot(x='Month', y='Cloud Cover Total', data=df, palette='Set2')

plt.title('Cloud Cover Total Over Time')
```

```python
plt.figure(figsize=(12, 8))

cloud_data = {
    'Month': df['Month'],
    'High Clouds': df['Cloud Cover High [high cld lay]'],
    'Medium Clouds': df['Cloud Cover Medium [mid cld lay]'],
    'Low Clouds': df['Cloud Cover Low [low cld lay]']
}
cloud_df = pd.DataFrame(cloud_data)

sns.barplot(x='Month', y='value', hue='variable', data=pd.melt(cloud_df, id_vars=['Month']), estimator=np.mean, palette='Set2')

plt.title('Cloud Cover at Different Levels Over Time')
plt.ylabel('Cloud Cover')
```

```python
sns.lineplot(x='Month', y='Relative Humidity [2 m]', data=df)

plt.title('Relative humidity graph at 2 meter level')
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
sns.barplot(x='Month', y='Snowfall Amount', data=df, palette='Blues')
plt.title('Total amount of snowfall')

plt.subplot(2, 1, 2)
sns.barplot(x='Month', y='Snow Depth', data=df, palette='Greens')
plt.title('Depth of snow')

plt.tight_layout()
```
